```python
df_act_g12 = df_act_ca[df_act_ca['AvgScrRead'] != '*']
df_act_g12.head()
```

```python
df_act_g12.reset_index(drop=True)
```

```python
df_act_g12.info()
```

```python
df_act_g12[['Enroll12', 'NumTstTakr', 'AvgScrRead', 'AvgScrEng',
       'AvgScrMath', 'AvgScrSci', 'NumGE21']] = df_act_g12[['Enroll12', 'NumTstTakr', 'AvgScrRead', 'AvgScrEng',
       'AvgScrMath', 'AvgScrSci', 'NumGE21']].astype('int')
```

```python
df_act_g12['PctGE21'] = df_act_g12['PctGE21'].astype('float')
```

```python
df_act_g12.info()
```

```python
df_act_g12_avg = df_act_g12[['CName', 'Enroll12', 'NumTstTakr', 'AvgScrRead', 'AvgScrEng','AvgScrMath', 'AvgScrSci', 'NumGE21']]
df_act_g12_avg = df_act_g12_avg.groupby('CName').sum()
df_act_g12_avg.head()
```

```python
df_act_g12_pct = df_act_g12[['CName','PctGE21']]
df_act_g12_pct = df_act_g12_pct.groupby('CName').sum()
df_act_g12_pct.head()
```

```python
df_act_g12_cname = pd.concat([df_act_g12_avg,df_act_g12_pct],axis=1)
df_act_g12_cname.head(10)
```

```python
df_act_g12_cname['Percent_Test_Takers'] = np.round((df_act_g12_cname['NumTstTakr']/df_act_g12_cname['Enroll12'])*100,2)
```

```python
df_act_g12_cname.head(10)
```

```python
### Top 10 of Percent of Test Takers Grade 12 in each State
sns.set_style('whitegrid')
df_act_g12_cname['Percent_Test_Takers'].sort_values(ascending=False)[:10].plot(kind='bar',figsize=(10,5),color='salmon',fontsize=12)
plt.title('G12 : Top 10 of Percent of Test Takers in each City\n',fontsize=15)
plt.xlabel('City Name')
plt.ylabel('Percent of Test Takers')
plt.xticks(rotation=30);
```

```python
df_act_g12_cname['Diff_Test'] = df_act_g12_cname['Enroll12'] - df_act_g12_cname['NumTstTakr']
```

```python
df_act_g12_cname[['Enroll12','NumTstTakr','Diff_Test']].sort_values(by = 'Enroll12',ascending=False)[:10].sort_values(by = 'Diff_Test' ,ascending=False)
```

```python
### Top 10 of Enrollment Grade 12 in each State
sns.set_style('whitegrid')
df_act_g12_cname[['Enroll12','NumTstTakr','Diff_Test']].sort_values(by = 'Enroll12',ascending=False)[:10].sort_values(by = 'Diff_Test' ,ascending=False).plot(kind='bar',figsize=(12,5),width=0.7,fontsize=12,color=['#1BE9B1','#1BE9B1','salmon'],alpha=0.8)
plt.title('G12 : Top 10 Number of student in each City\n',fontsize=15)
plt.legend(['Student','Test_Takers','Diff'],loc=10)
plt.xlabel('City Name')
plt.ylabel('Number of student')
plt.xticks(rotation=30);
```

```python
df_act_g12_cname['AvgTotalScr'] = df_act_g12_cname[['AvgScrRead','AvgScrEng','AvgScrMath','AvgScrSci']].apply(lambda x:x.mean(),axis=1)
df_act_g12_cname.head()
```

```python
df_act_g12_cname_top10 = df_act_g12_cname.sort_values(by ='Percent_Test_Takers', ascending=False)[:10]
df_act_g12_cname_top10.head(10)
```

```python
avg_per_top10_g12 = df_act_g12_cname_top10.PctGE21.mean()
avg_per_top10_g12
```

```python
df_act_g12_cname_top10[['PctGE21']].plot(kind='bar',figsize=(12,5),fontsize=10,color='#F2B520',alpha=0.8)
plt.axhline(avg_per_top10_g12,color='r',lw=2,linestyle='--',alpha=0.6,label='Mean of PctGE21')
plt.ylabel('Percent of Test Takers whose ACT Composite Scores are Greater or Equal to 21',fontsize=8)
plt.xlabel('City Name')
plt.xticks(rotation=30);
plt.title('G12 : Summary Top 10 of Test Takers(%) by ACT score in California\n',fontsize=15)
plt.legend();
```
