## Visualization of score

```python
df_sat.head()
```

### EBRW score in each year

```python
### Plot histogram for visualization of distribution of EBRW score in 2017

mean_EBRW_2017 = df_sat['ebrw_2017'].mean()
sd_EBRW_2017 = df_sat['ebrw_2017'].std()
sns.distplot(df_sat['ebrw_2017'],bins=20,color = '#EC7063')
plt.axvline(mean_EBRW_2017,linewidth=3,linestyle='--')
plt.title(f'EBRW in 2017 (mean = {mean_EBRW_2017:.2f}, sd = {sd_EBRW_2017:.2f})\n')
plt.grid()
```

```python
### Plot histogram for visualization of distribution of EBRW score in 2018

mean_EBRW_2018 = df_sat['ebrw_2018'].mean()
sd_EBRW_2018 = df_sat['ebrw_2018'].std()
sns.distplot(df_sat['ebrw_2018'],bins=20,color = '#D4AC0D')
plt.axvline(mean_EBRW_2018,linewidth=3,linestyle='--')
plt.title(f'EBRW in 2018 (mean = {mean_EBRW_2018:.2f}, sd = {sd_EBRW_2018:.2f})\n')
plt.grid()
```

```python
### Plot histogram for visualization of distribution of EBRW score in 2019
### Compare with EBRW mean score of all State and EBRW score of California
### Show that EBRW score of California lessthan EBRW mean score of all State

mean_EBRW_2019 = df_sat['ebrw_2019'].mean()
sd_EBRW_2019 = df_sat['ebrw_2019'].std()
sns.distplot(df_sat['ebrw_2019'],bins=20,color = '#148F77')

plt.axvline(mean_EBRW_2019,linewidth=3,linestyle='--',label='mean')
plt.axvline(df_sat.loc['California']['ebrw_2019'].max(),
            linewidth=3,linestyle='--',label='California',color='r')

plt.title(f'EBRW in 2019 (mean = {mean_EBRW_2019:.2f}, sd = {sd_EBRW_2019:.2f})\n')
plt.legend()
plt.grid()
```

```python
### Compare EBRW score in 3 Years

sns.distplot(df_sat['ebrw_2017'],bins=20,color = '#EC7063',hist=False,label='2017')
sns.distplot(df_sat['ebrw_2018'],bins=20,color = '#D4AC0D',hist=False,label='2018')
sns.distplot(df_sat['ebrw_2019'],bins=20,color = '#148F77',hist=False,label='2019')
plt.axvline((df_sat.ebrw_2017.mean()+df_sat.ebrw_2018.mean()+df_sat.ebrw_2019.mean())/3,linewidth=3,linestyle='--',label='mean')
plt.xlabel('EBRW Score')
plt.legend()
plt.title('Comparative EBRW in 3 Years\n');
```
